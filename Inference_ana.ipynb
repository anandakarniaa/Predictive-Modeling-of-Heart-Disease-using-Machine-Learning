```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
import joblib  # jika ingin menyimpan model
```

```python
# Contoh data baru
# data_baru = pd.DataFrame([{
#     'BMI': 28.5,
#     'Smoking': 'Yes',
#     'AlcoholDrinking': 'No',
#     'Stroke': 'No',
#     'PhysicalHealth': 3.0,
#     'MentalHealth': 5.0,
#     'DiffWalking': 'No',
#     'Sex': 'Female',
#     'AgeCategory': '55-59',
#     'Race': 'White',
#     'Diabetic': 'No',
#     'PhysicalActivity': 'Yes',
#     'GenHealth': 'Good',
#     'SleepTime': 6.0,
#     'Asthma': 'No',
#     'KidneyDisease': 'No',
#     'SkinCancer': 'No'
# }])


data_baru = pd.DataFrame([{
    'BMI': 28.5,
    'Smoking': 'Yes',
    'AlcoholDrinking': 'No',
    'Stroke': 'No',
    'PhysicalHealth': 3.0,
    'MentalHealth': 5.0,
    'DiffWalking': 'No',
    'Sex': 'Female',
    'AgeCategory': '55-59',
    'Race': 'White',
    'Diabetic': 'No',
    'PhysicalActivity': 'Yes',
    'GenHealth': 'Good',
    'SleepTime': 6.0,
    'Asthma': 'No',
    'KidneyDisease': 'No',
    'SkinCancer': 'No'
}])
```

```python
# Fitur numerik dan kategorikal
num_features = ['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']
cat_features = [col for col in data_baru.columns if col not in num_features]
```

```python
# Preprocessing
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_features)
])
```

```python
# Load model terbaik
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
```

```python
# Pipeline
pipeline = Pipeline([
    ('preprocess', preprocessor),
    ('classifier', model)
])
```
